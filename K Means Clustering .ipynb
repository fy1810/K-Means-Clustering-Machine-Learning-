```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('student_marks.csv')
df
```

```python
df.head(4)
```

```python
df.tail()
```

```python
marks=df.iloc[:,1:4]
marks
```

```python
distortion=[]
distortion
```

```python
k=range(1,13)
k
```

```python
for i in k:
    kmeansmodel=KMeans(n_clusters=i,max_iter=25)
    kmeansmodel.fit(marks)
    distortion.append(kmeansmodel.inertia_)
```

```python
plt.plot(figure=(13,0))
plt.plot(k,distortion,'bo-')
plt.xlabel('Hoof Clusters')
plt.ylabel('WSS')
plt.title('KMeans Clustering')
plt.show()
```

```python
kmeandmodelfinal=KMeans(n_clusters=3,max_iter=25)
kmeandmodelfinal.fit(marks)
```
